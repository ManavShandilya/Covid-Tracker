```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv("covid_vaccine_statewise.csv")
```

```python
print(df.describe())
print(df.head(10))
```

```python
first_dose_data = df.groupby('State')['First Dose Administered'].max().reset_index()
print('First Dose:\n', first_dose_data)
```

```python
# Number of persons state-wise vaccinated for the second dose in India
second_dose_data = df.groupby('State')['Second Dose Administered'].sum()
print('Second Dose:\n', second_dose_data)
```

```python
# Number of Males vaccinated
males_vaccinated = df['Male(Individuals Vaccinated)'].sum()
print('Males Vaccinated:', males_vaccinated)
```

```python
females_vaccinated = df['Female(Individuals Vaccinated)'].sum()
print('Females Vaccinated:', females_vaccinated)
```

```python
sns.boxplot(x = 'CoviShield',data = df)
plt.show()
```

```python
# removing outliers
def Remove_outlier(df,var):
	final_df = []
	q1 = df[var].quantile(0.25)
	q3 = df[var].quantile(0.75)
	iqr = q3-q1
	mini=q1-1.5*iqr
	maxi=q3+1.5*iqr
	for j in df[var]:
		if(j<mini or j>maxi):
			final_df.append(j)
			df.drop(df.loc[df[var]==j].index, inplace=True)
```

```python
Remove_outlier(df,"CoviShield")
# boxplot after removing outliers
plt.title('Boxplot After Removing Outliers')
sns.boxplot(x = 'CoviShield',data = df)
plt.show()
```

```python
plt.bar(first_dose_data['State'], first_dose_data['First Dose Administered'], label='First Dose')
plt.title('Number of Persons Vaccinated for First Dose in Each State')
plt.xlabel('State')
plt.ylabel('Number of Persons')

# Rotate the x-axis labels for better visibility
plt.xticks(rotation=85)

# Display the plot
plt.show()
```

```python
# Min-max scaling is a normalization technique used to transform features to a common scale.
# In machine learning, it is often essential to normalize the features to a range between 0 and 1.

# Initialize the scaler
scaler = MinMaxScaler()

# Apply min-max normalization on the 'First Dose Administered' column
first_dose_data['First Dose Administered'] = scaler.fit_transform(first_dose_data[['First Dose Administered']])

# Plot the bar chart with normalized values
plt.bar(first_dose_data['State'], first_dose_data['First Dose Administered'], label='First Dose')

# Set the x-axis label, y-axis label and title of the plot
plt.xlabel('State')
plt.ylabel('First Dose Administered (Normalized)')
plt.title('Number of persons state-wise vaccinated for the first dose in India (Normalized)')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=90)

# Show the legend and plot the chart
# plt.legend()
plt.show()

```
